```python
import tkinter as tk
from bs4 import BeautifulSoup 
import pandas as pd
import requests as rq
import matplotlib.pyplot as plt
from tkinter import *
```

```python
URL = 'https://www.worldometers.info/coronavirus/country/us/'
page = rq.get(URL)
html = page.text
soup = BeautifulSoup(html,'html.parser')
#print(soup.title.text)
#livedata = soup.find_all('div',id='maincounter-wrap')

#for i in livedata:
#    print(i.text)
    
tb = soup.find('tbody')
rows = tb.find_all('tr')
states = []
cases = []
newcases = []
deaths = []
newdeaths = []

for i in rows:
    td = i.find_all('td')
    states.append(td[0].text)
    cases.append(td[1].text)
    newcases.append(td[2].text)
    deaths.append(td[3].text)
    newdeaths.append(td[4].text)
    
# \n remover
    fs =[]
    fn=[]
    fd=[]
    for i in states:
        fs.append(i.strip())
    for j in newcases:
        fn.append(j.strip())
    for k in deaths:
        fd.append(k.strip())
    
head = ['State','Cases','New Cases','Deaths',"   Today's Deaths"]
df = pd.DataFrame(list(zip(fs,cases,fn,fd,newdeaths)),columns=head)
# uncomment to save it as csv df.to_csv('COVIData.csv')
#print(df)
```
